

*   **Financial Metrics:** The movie industry is a winner-take-all market. Most movies (the "long tail") are produced and earn very little, while a handful of large-scale productions account for the majority of the total budget spent and revenue generated. Analyzing the median (which is less affected by skew) would provide a more realistic "typical" movie value than the mean.

*   **Rating Metrics:** Unlike financial data, ratings follow a more standard distribution. Most movies are neither critical masterpieces nor total disasters. There is an inherent ceiling and floor to ratings, compressing the distribution compared to unconstrained financial figures.

*   **Voting Metrics:** The number of votes acts as a proxy for movie popularity and exposure. A movie with a low vote count may have a high/low rating, but its popularity is low. Only a tiny fraction of movies achieve widespread audience engagement necessary to generate hundreds of thousands of votes.

```python
categorical_columns = ['Genre', 'Country', 'Director', 'LeadActor']
```

```python
for cols in categorical_columns:
  df[cols].value_counts().head(20).plot(kind="bar")
  plt.title(f"Top 20 {cols} counts")
  plt.show()
```

* **Genre:-**	Dominant Genres: Drama, Comedy, Action and Thriller are the most frequent. This reflects the high production volume of genre films that appeal to mass audiences and the critical mass of drama films across all eras.
* **Country:-**	USA Dominance: The USA overwhelmingly top the list, followed by countries like the UK, Canada, and India (Bollywood). This confirms the global production dominance of Hollywood and American filmmaking.
* **Director:-**	Prolific Filmmakers: A handful of highly productive directors appear in the Top 20. The steep drop-off between the top director and the 20th shows that production volume is concentrated among a select group of professionals.
* **Lead Actor:-**	Star Power/Consistency: Similar to directors, this plot identifies the most prolific lead actors. These actors have either been in the industry for a long time or are consistently cast as leads, indicating a concentration of roles among a few "bankable" stars.

```python
#RELATION BETWEEN BOX OFFICE AND BUDGET
sns.scatterplot(data = df, x='BudgetUSD', y='Global_BoxOfficeUSD', alpha=0.3)
plt.title("Budget vs Global Box office")
plt.xscale("log")
plt.yscale("log")
plt.show()
```

* **Trend**: A clear positive linear relationship is visible on the log-log scale.

* **Analysis**: This confirms that money matters; a higher budget generally correlates with higher box office revenue. The budget facilitates better production values, marketing spend, and star hiring, all contributing to potential success.

* **Outliers**:

  * High ROI (Above the trend line): These points represent "sleeper hits" or mega-hits with modest budgets (e.g., a horror film or a low-budget independent hit) that returned a high profit multiple.

  * Low ROI (Below the trend line): These are expensive blockbusters that failed to recoup costs (i.e., "bombs"). They are high-budget movies that underperformed significantly.

```python
#ANALYSIS OF RATINGS
sns.scatterplot(data = df, x='IMDbRating', y='RottenTomatoesScore', alpha=0.3)
plt.title("IMDb vs Rotten Tomatoes Ratings")
plt.show()
```

* **Trend**: A moderate to strong positive correlation (e.g., $r > 0.6$).

* **Analysis**: This shows a general agreement between critics (Rotten Tomatoes) and the general audience (IMDb). Movies that critics like, audiences tend to like, and vice versa.

* **Disagreement/Outliers**:
  * High RT, Low IMDb: Films rated highly by critics but poorly by the audience. Often includes art-house, niche, or esoteric films.  
  * Low RT, High IMDb: Films rated poorly by critics but highly by the audience. These are typically popcorn blockbusters, franchise films, or highly emotional dramas that critics dismiss as formulaic but that audiences enjoy.

```python
genre_year_stats = df.groupby(["Genre", "ReleaseYear"]).agg({
    "Global_BoxOfficeUSD" : "mean",
    "BudgetUSD" : "mean",
    "IMDbRating" : "mean",
    "RottenTomatoesScore" : "mean"
}).reset_index()
```

```python
genre_year_stats
```

```python
plt.figure(figsize=(15, 10))
sns.lineplot(data = genre_year_stats, x="ReleaseYear", y="Global_BoxOfficeUSD", hue="Genre")
plt.title("Average Global Box Office by Genre over the years")
```

* **Trend**: All genre lines show a clear, consistent upward trajectory from the mid-20th century to the present.

* **Analysis**: This overall rise is primarily driven by inflation (movie tickets and budgets are more expensive over time) and the expansion of the global market (more theaters, higher population).

* **Genre Hierarchy**: The lines for Action, Adventure, and Sci-Fi will consistently be positioned at the top of the chart, showing that these genres have the highest average financial return. This is due to their inherent suitability for large-scale, high-budget productions aimed at the global market. **Drama** and **Comedy** lines will be lower, showing a lower average return, despite their high production volume.

```python
top_actors = df.groupby("LeadActor")["IMDbRating"].mean().sort_values(ascending=False).head(10)
top_actors
```

```python
top_actors.plot(kind="bar")
plt.title("Top 10 actors by average IMDb Rating")
plt.show()
```

* **Trend**: This plot identifies actors who consistently appear in highly-rated films, not necessarily the most popular or prolific ones (as seen in the Top 20 count plot).

* **Analysis**: The actors featured here often select roles in critically acclaimed, independent, or prestige projects, indicating a focus on quality over sheer volume.

* **Insight**: This metric suggests a potential correlation between certain actors and a higher likelihood of a film's critical success, which can be a valuable insight for studios looking to attach talent to higher-quality, award-contending projects. The range of the average ratings will be quite narrow (e.g., $7.5$ to $8.0$), highlighting the high bar for consistency required to make this list.
