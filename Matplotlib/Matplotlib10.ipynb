```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/pokemon.csv")

type_count = df["Type1"].value_counts(ascending = True)

plt.barh(type_count.index, type_count, color="lightgreen",
                                       edgecolor = "black")

plt.title("No. of Pokemon by Primary type")
plt.xlabel("Count")
plt.ylabel("Type")

plt.tight_layout()
plt.show()
```
